```python
color_banana = ["#e41a1c", "#377eb8", "#4daf4a", "#984ea3", "#ff7f00", "#ffff33", "#a65628"]
sns.set_palette(color_banana)
sns.palplot(sns.color_palette())
```

```python
df_vineyard_box = pd.read_csv("fire progression -- 633 buffer _ urban_to_TL2 - vegetation_types_633_box (5).csv")
sns.set(style="whitegrid")
plt.figure(figsize=(14,10))
#plt.style.use('classic')
#fig, ax = plt.subplots()
#plt.title('Vineyard/Orchard Fire Buffers Comparison')
sns_plot = sns.boxplot(x="Time since fire ignition (hours)",y="Urban area burned (square meters)", 
                  hue="Buffer type", data=df_vineyard_box, linewidth=3.5, palette=color_banana, order=range(17, 39, 3))

#plt.suptitle("3(c). Vineyard Fire Buffers Comparison")
#sns.swarmplot(x="time since fire ignited(hour(s))", y="result urban burnt(m^2)", data=df_vineyard_box, color=".25", ax=ax)
#sns.lmplot(x="Time since fire ignition (hours)",y="Urban area burned (square meters)", hue="buffer type", data=df_vineyard_box)
#sns_plot.axes.set_title("Vineyard/Orchard Fire Buffers Comparison",fontsize=30)
sns_plot.set_xlabel("Time since fire ignition (hours)",fontsize=25)
sns_plot.set_ylabel("Urban area burned $(m^2)$",fontsize=25)
sns_plot.tick_params(labelsize=20)
plt.legend(fontsize=25, title_fontsize=25)

plt.show()
sns_plot.figure.savefig("figure2c.pdf")
sns_plot.figure.savefig("2c.png")
```
